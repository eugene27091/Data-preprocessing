```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic') 
```

```python
for brand in df['brand']:
   brand_kind = df['brand'].unique()

print(brand_kind)
```

```python
plt.figure(figsize=(20,10))
plt.title("강아지 사료 인기 브랜드 판매 추이", fontsize=15)
for brand in brand_kind:
    d_ = df[(df["brand"]==brand) & (df["date"].dt.year >= 2015)]
    plt.plot(d_["date"], d_["num"], "-", label=brand, alpha=.6)
plt.grid()
plt.legend(fontsize=13)
plt.xticks(rotation=90)
plt.show()
```
